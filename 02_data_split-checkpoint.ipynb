```python
# ----------------------------------------------------------
# ğŸ“˜ Split dataset into train/test and save test sample
# ----------------------------------------------------------

import pandas as pd
from sklearn.model_selection import train_test_split

# 1ï¸âƒ£ Load your full dataset (the one you used to train)
df = pd.read_csv("data/card_fraud.csv")   # <-- adjust path if needed

# 2ï¸âƒ£ Split 80â€“20 (stratify ensures same fraud ratio in both)
train_df, test_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["isFraud"],  # make sure column name matches your dataset
    random_state=42
)

# 3ï¸âƒ£ Save both (optional)
train_df.to_csv("data/card_fraud_train.csv", index=False)
test_df.to_csv("data/card_fraud_test.csv", index=False)

print(f"âœ… Train shape: {train_df.shape}, Test shape: {test_df.shape}")
print("ğŸ’¾ Files saved:")
print(" - data/card_fraud_train.csv")
print(" - data/card_fraud_test.csv")
```
